```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
os.getcwd()
```

```python
df = pd.read_csv('../RedbackRaceData/PressureToTempAndLap/PressureToTempFront.csv')
```

```python
# df.plot.scatter(x='AvgTempFront', y='Front_Pressure')
plt.title('Front Tyre Pressure to Average Core Temp Front')
ax = sns.regplot(x='AvgTempFront', y='Front_Pressure', data=df, ci=None, line_kws={"color": "orange"})
```
